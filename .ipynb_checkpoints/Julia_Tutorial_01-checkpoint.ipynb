```julia
a = rand(["a","b","NA"], 20)
a2 = Array{Union{Missing, String}}(a)
```

```julia
for i in eachindex(a2)
    #i == "NA" && (as[1] = missing)
    if a2[i] == "NA"
        a2[i] = missing
    end
end
```

```julia
a3 = [x == "NA" ? missing : x for x in, a] # ternary operatore syntax
```

```julia
#a4 = map(x -> if x == "NA"; missing; else x; end, a)
function convertna(x) 
    if x == "NA"
        return missing
    end 
    return x
end
```
